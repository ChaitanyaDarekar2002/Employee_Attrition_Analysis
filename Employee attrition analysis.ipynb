```python
for col in ['Department','JobRole','Attrition']:
    print(col)
    print(data[col].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

data['Attrition']=le.fit_transform(data['Attrition'])
```

```python
data['Attrition'].value_counts().plot(kind='bar')
```

```python
pd.crosstab(
    data['Gender'],
    data['Attrition']
).plot(kind='bar')

plt.title("Attrition by Gender")
plt.show()
```

```python
pd.crosstab(
    data['Department'],data['Attrition']
).plot(kind='line')
```

```python
plt.hist(data['Age'])
```

```python
plt.hist(data['MonthlyIncome'])
plt.title("Monthly Income Distribution")
plt.show()
```

```python
pd.crosstab(
    data['OverTime'],
    data['Attrition']
).plot(kind='bar')
plt.title("Overtime vs Attrition")
plt.show()
```

```python
corr=data.corr(numeric_only=True)
plt.figure(figsize=(15,8))
plt.imshow(corr)
plt.colorbar()
plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)
plt.yticks(
    range(len(corr.columns)),
    corr.columns
)
plt.show()
```

```python
pd.crosstab(
    data['WorkLifeBalance'],
    data['Attrition']
).plot(kind='bar')
plt.title("Work-Life Balance vs Attrition")
plt.show()
```

```python
pd.crosstab(
    data['JobSatisfaction'],
    data['Attrition']
).plot(kind='bar')
plt.title("Job Satisfaction vs Attrition")
plt.show()
```
